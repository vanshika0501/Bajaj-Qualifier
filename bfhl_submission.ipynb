```python
import requests


name = "Vanshika Saraf"
registration_number = "0827AL221141"
email_address = "vanshikasaraf@gmail.com"

payload = {
    "name": name,
    "regNo": registration_number,
    "email": email_address
}

print("Requesting webhook generation...")


webhook_response = requests.post(
    url="https://bfhldevapigw.healthrx.co.in/hiring/generateWebhook/PYTHON",
    json=payload
)

if webhook_response.ok:
    response_data = webhook_response.json()
    webhook_link = response_data.get("webhook")
    token = response_data.get("accessToken")
    
    print("Received Webhook URL:", webhook_link)
    print("Received Access Token:", token)
else:
    print("Error: Unable to generate webhook.")
    print("HTTP Status:", webhook_response.status_code)
    print("Details:", webhook_response.text)
    exit(1)


last_number = int(registration_number[-1])
print("\nBased on your registration number...")

if last_number % 2 != 0:
    print("Assigned SQL QUESTION 1:")
    print("https://drive.google.com/file/d/1q8F8g0EpyNzd5BWk-voe5CKbsxoskJWY/view?usp=sharing")
else:
    print("Assigned SQL QUESTION 2:")
    print("https://drive.google.com/file/d/1_PO_1ZvmDqAZJv77XRYsVben11Wp2HVb/view?usp=sharing")


sql_submission_query = """
SELECT 
    p.AMOUNT AS SALARY,
    CONCAT(e.FIRST_NAME, ' ', e.LAST_NAME) AS NAME,
    TIMESTAMPDIFF(YEAR, e.DOB, CURDATE()) AS AGE,
    d.DEPARTMENT_NAME
FROM PAYMENTS p
JOIN EMPLOYEE e ON p.EMP_ID = e.EMP_ID
JOIN DEPARTMENT d ON e.DEPARTMENT = d.DEPARTMENT_ID
WHERE DAY(p.PAYMENT_TIME) != 1
ORDER BY p.AMOUNT DESC
LIMIT 1;
""".strip()


headers = {
    "Authorization": token,
    "Content-Type": "application/json"
}

submission_payload = {
    "finalQuery": sql_submission_query
}

print("\nSending your SQL solution...")

final_response = requests.post(
    url=webhook_link,
    headers=headers,
    json=submission_payload
)

print("Response Code:", final_response.status_code)
print("Response Message:", final_response.text)

if final_response.status_code == 200:
    print(" Query submitted successfully!")
else:
    print("Submission failed.")
```
